```python
import numpy as np
import os
import _pickle as pickle
from scipy import spatial
from gensim.models import Word2Vec
import logging
import json
import regex as re
import random
import scipy
```

Preprocessing 

```python
######  function to extract text from documents     #########
def extract_text(filename):
    '''
    Extract the text from the file name (json file) and
    index the content from paper_id, title, abstract and body_text fields
    Retuns - text of title, abstract and bodt_text
    '''

    file = open('crps1\pdf_json\\'+filename)
    body_text = ""
    abstract = ""
    title = ""
    paper_id = ""

    paper_content = json.load(file)

    if 'paper_id' in paper_content:
        paper_id = paper_content['paper_id']
        
    if 'title' in paper_content['metadata']:
        title = paper_content['metadata']['title']

    if 'abstract' in paper_content:
        for abs in paper_content['abstract']:
            abstract = abstract + abs['text']
    if 'body_text' in paper_content:
        for bt in paper_content['body_text']:
            body_text = body_text + bt['text']
    resultant=''
    if abstract!='':
        resultant+=abstract+'. '
    if body_text!='':
        resultant+=body_text
    return resultant

def remove_non_ascii_characters(text:str):      
    resultant_text=''
    for i in text:
        x=ord(i)
        if (x>31 and x<127):
            resultant_text+=i
    return resultant_text


def preprocess(text:str):
    text=re.split('[!?.]',remove_non_ascii_characters(text).lower())
    return text
    
```

```python
NO_OF_FILES=1500        #select number of files to work on
```

```python
SENTENCE_UPPER_LIMIT=20             #setting the length of sentence to work on
SENTENCE_LOWER_LIMIT=8                 # we are going to build sentences vectors for only those sentences
                                        # whose length is in between 8 and 20
```

```python
stream = os.popen('dir /b crps1\pdf_json ')     #applying preprocessing on selected files
files=stream.read().split()
count=0
fil=open(path+'sentences.txt','w')
for f in files:
    data=extract_text(f).lower()
    data=preprocess(data)
    for sent in data:
        try:
            len_sent=len(sent.split())      
            if len_sent>SENTENCE_LOWER_LIMIT and len_sent<SENTENCE_UPPER_LIMIT:
                fil.writelines(sent+'\n')           #writing the senetnces inside a text file named 'sentences.txt'
        except:
            pass
    count+=1
    print(count,'many files processed')
    if count>=NO_OF_FILES:
        break
fil.close()
```

```python
f=open(path+'sentences.txt','r')
sentences=f.readlines()         # readinmg all the sentences
f.close()
```

```python
WORD_VEC_SIZE=100       #length of word vectors
SENTENCE_VEC_SIZE=1000  #length of sentence vector
OPTION=0   #could be any of {0,1,2}
            # where 0 means sentence vector will be created from both direction and then concatinated
            # 1 means sentence vector will be created only with from left to right direction
            # 2 means sentence vector will be created only with from right to left direction
```

```python
splited_sentences=[sentence.split()   for sentence in sentences]        #making list of list of words  
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
model = Word2Vec(vector_size=WORD_VEC_SIZE, window=5, workers=5,min_count=1) #initializing the word2vec model
model.build_vocab(splited_sentences)
model.train(splited_sentences,total_examples=model.corpus_count,epochs=20)  # creating the word vector with gensim word2vec model
```

```python
model.save(path+'trainedModel_4.bin')  #saving the model for future access 
```

```python
u=np.random.normal(0,0.01,(SENTENCE_VEC_SIZE,SENTENCE_VEC_SIZE))        #initializing u,h0 and w
h0=np.random.normal(0,0.01,SENTENCE_VEC_SIZE)
w=np.random.normal(0,0.01,(SENTENCE_VEC_SIZE,WORD_VEC_SIZE))
```

```python
SENTENCE_COUNT=len(splited_sentences)


#initializing matrix to store sentence vector
if OPTION==0:
    sentence_vectors=np.zeros(shape=(SENTENCE_COUNT,SENTENCE_VEC_SIZE*2))   
elif OPTION==1 or OPTION==2:
    sentence_vectors=np.zeros(shape=(SENTENCE_COUNT,SENTENCE_VEC_SIZE))

h_i_rev=u.dot(h0)
h_i_fwd=u.dot(h0)

```

```python
#creating sentence vector
for pos,sentence in enumerate(splited_sentences):   #going through every sentence in the sentence list
    if OPTION==0 or OPTION==1:
        for word in sentence:                       # picking each word of the current sentence 
            y=scipy.stats.zscore(model.wv[word])
            h_iP1_fwd=np.tanh(np.dot(w,y)+np.dot(u,h_i_fwd))    #and calculating h[i+1]
            if OPTION==0:
                sentence_vectors[pos]+=np.hstack((h_iP1_fwd,np.zeros(shape=(SENTENCE_VEC_SIZE))))   #for both way operation we are just adding the vector for forward part .later the vector for reverse part will be added
            else:
                 sentence_vectors[pos]+=h_iP1_fwd       #for forward operation only adding the vector created 
            h_i_fwd=h_iP1_fwd/SENTENCE_VEC_SIZE

    elif OPTION==0 or OPTION==2:
        sentence_rev=sentence.reverse()
        for word in sentence_rev:
            y=scipy.stats.zscore(model.wv[word])
            h_ip1_rev=np.tanh(np.dot(w,y)+np.dot(u,h_i_rev))
            if OPTION==0:
                sentence_vectors[pos]+=np.hstack((np.zeros(shape=(SENTENCE_VEC_SIZE)),h_ip1_rev))   #adding the reverse 
            else:
                 sentence_vectors[pos]+=h_ip1_rev
            h_i_rev=h_ip1_rev/SENTENCE_VEC_SIZE

#so for both way operation our vector will be of size 2*sentence_vec_size( 1st half will contain the forward vector and second half will contain the reverse vetor part)
```

```python
def findSimilar(x,n):   #function for finding n most similar sentences for a given sentence
  vec=sentence_vectors[x]
  vecList=[]
  for i in range(sentence_vectors.shape[0]):
    if i==x:
      continue
    vecList.append([sentences[i].strip(),1-spatial.distance.cosine(vec,sentence_vectors[i])])
  vecList=sorted(vecList,key= lambda x: x[1],reverse=True)
  if n<=len(vecList):
    return vecList[:n]
  else:
    return vecList

def print_similar_sentences(x,n):
  similar_sentences=findSimilar(x,n)
  print('sentences similar to the sentence  "',sentences[x],'" at position: ',x,' are :-')
  for i in similar_sentences:
    print(i)
```

```python
for i in range(5):
    print(i+1,'>')
    choosen_sentence_index=random.randint(0,SENTENCE_COUNT)
    print_similar_sentences(choosen_sentence_index,5)
    print('\n')
```
